```python
from neo4j import GraphDatabase
from random import randint,choice
import datetime
```

```python
conn = GraphDatabase.driver(uri = "bolt://localhost:7687",auth=("vishal",'123456'))
```

```python
sess = conn.session(database="debitcardtran")
```

```python
ATM_Id_List = ["0407F","123ON","9099A","7833E","E234OP","4895L","2745K","2280Q","1099T","4489K",'RT44Y','WE23G','BZ679','QP23K','U99OP']
```

```python
%%time
for i in range(1,200):
    acct_num = randint(4370223011,9370223011)
    card_num = randint(4370223011119990,9370223011119990)
    last4digit = str(card_num)[12:16]
    #print(acct_num, card_num)
    sess.run("MERGE(a:Account{AcctNum:$acct_num}) \
              MERGE(c:DebitCard{DebitCardNum :$card_num}) \
              MERGE(l:Last4Digit{DebitCardLast4Digit : $last4digit}) \
              CREATE (a)-[:HAS_DEBITCARD]->(c)\
              CREATE (c)-[:HAS_LAST4DIGIT]->(l)\
             ",
            acct_num = acct_num,
            card_num = card_num,
            last4digit = last4digit
            )
    itrtn = randint(1,100)
    print('Number of Iterations in loop # : '+ str(i) + ' are : '+ str(itrtn))
    for i in range(1,itrtn):
        start_date = datetime.date(2020, 1, 1)
        random_days = randint(1,364)
        Tran_dt = start_date + datetime.timedelta(days=random_days)
        #print(Tran_dt)
        Tran_amt=randint(1,1000)
        #print(Tran_amt)
        ATM_Id = choice(ATM_Id_List)
        Loc_Cd=randint(437022301,937022301)
        #print(Loc_Cd)
        sess.run("MATCH(c:DebitCard{DebitCardNum :$card_num}) \
                  MERGE(t:Transaction{TranAmt:$Tran_amt})\
                  MERGE(l:Location{Loc_Cd:$Loc_Cd})\
                  MERGE(d:TranDate{TranDt:$Tran_dt})\
                  MERGE(a:ATM{ATM_Id:$ATM_Id})\
                  CREATE (c)-[:MADE_TRAN_OF_AMT]->(t)\
                  CREATE (t)-[:TRAN_MADE_AT_LOC]->(l)\
                  CREATE (t)-[:TRAN_MADE_AT_ATM]->(a)\
                  CREATE (t)-[:TRAN_MADE_ON_DT]->(d)\
                 ",
                card_num = card_num,
                Tran_amt = Tran_amt,
                Loc_Cd = Loc_Cd,
                Tran_dt = Tran_dt,
                ATM_Id = ATM_Id
                )
```
